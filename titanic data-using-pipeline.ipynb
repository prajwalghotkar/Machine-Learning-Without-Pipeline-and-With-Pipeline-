# Let's Start Prajwal 

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -----> train-test-split

X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
y_train.sample()
```

```python
# Imputation Transfomer

trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# one hot encoding

trf2 = ColumnTransformer([
    ('one_hot_encoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1, 6])
], remainder='passthrough')
```

```python
# Scaling 

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature selection

trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model

trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline VS make_pipeline

#### Pipeline requires naming of steps, make_pipeline does not.
#### (Same applies to ColumnTransformer vs make_column_transformeer)

```python
# train
pipe.fit(X_train,y_train)
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# Cross validation using cross_validation_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train,y_train,cv=5,scoring='accuracy').mean()
```
